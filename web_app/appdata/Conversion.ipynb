## Generate a list of options for cities name in an object that can be implemented as a dropdown in Dash

```python
import pandas as pd
```

```python
df = pd.read_csv("cities.csv")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
options = list()
```

```python
for _,x in df.iterrows():
    city = x.city
    lat = str(x.lat)
    lon = str(x.lon)
    latlon = lat+" "+lon
    options.append({"label":city, "value":latlon})
```

```python
import pickle
```

```python
pickle.dump(options, open("options.pickle", 'wb'))
```
